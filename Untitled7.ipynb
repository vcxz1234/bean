```
import random
class Turtle(object):
    def __init__(self):
        self.x = random.randint(1, 10)
        self.y = random.randint(1, 10)
        self.power = 100
    def move(self):
        move_skill = [-2, -1, 0, 1, 2]
        new_x = self.x + random.choice(move_skill)
        new_y = self.y + random.choice(move_skill)
        self.x = self.is_vaild(new_x)
        self.y = self.is_vaild(new_y)
        self.power -= 1
    def is_vaild(self, value):
        if 1 <= value <= 10:
            return value
        elif value < 1:
            return abs(value)
        else:
            return 10 - (value - 10)
    def eat(self):
        self.power += 20
class Fish(object):
    def __init__(self):
        self.x = random.randint(1, 10)
        self.y = random.randint(1, 10)
    def move(self):
        move_skill = [-1, 0, 1]
        new_x = self.x + random.choice(move_skill)
        new_y = self.y + random.choice(move_skill)
        self.x = self.is_vaild(new_x)
        self.y = self.is_vaild(new_y)
    def is_vaild(self, value):
        if 1 <= value <= 10:
            return value
        elif value < 1:
            return abs(value)
        else:
            return 10 - (value - 10)
def main():
    turtle = Turtle()
    print(turtle.x, turtle.y)
    fishs = [Fish() for i in range(10)]
    while True:
        if turtle.power <= 0:
            print("烏龜沒有體力了，魚獲勝")
            break
        elif len(fishs) == 0:
            print("魚被吃光，烏龜獲勝")
            break
        else:
            turtle.move()
            for fish in fishs:
                fish.move()
                if turtle.x == fish.x and turtle.y == fish.y:
                    turtle.eat()
                    fishs.remove(fish)
                    print("魚被烏龜吃掉， 還剩%d條魚....." % (len(fishs)))
                    print("烏龜的體力為%s" % (turtle.power))
            else:
                print("烏龜沒有吃到魚，體力為%s" % (turtle.power))
if __name__ == "__main__":
    print("遊戲開始".center(30, "*"))
    main()
```
